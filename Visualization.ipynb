```python
import pandas as pd
from pylab import arrow, xlim, ylim
import numpy as np
import matplotlib.pyplot as plt
```

```python
pre = pd.read_csv('pre.csv')
res = pd.read_csv('res.csv')
```

```python
pre = np.array(pre)
res = np.array(res)
```

```python
for j in range(20):
    for i in range(1+j*12, 12+j*12):
        arrow1 = plt.arrow(pre[1][i], pre[2][i],pre[1][i+1] - pre[1][i], pre[2][i+1] - pre[2][i], head_width=0.01, color = 'r', label = 'prediction')
        arrow2 = plt.arrow(res[1][i], res[2][i],res[1][i+1] - res[1][i], res[2][i+1] - res[2][i], head_width=0.01, color = 'g', label='target')
        xlim(0.7, 1.7)
        ylim(2.0, 3.0)
    plt.legend([arrow2,arrow1], ['Target','Prediction'])
    plt.show()
```
